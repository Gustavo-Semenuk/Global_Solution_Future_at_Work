## Validation dados por Estado

```sql
SELECT 
a.sigla_uf,
a.avg_ibc,
a.avg_crescimento,
b.ano,
b.idhm,
b.indice_escolaridade,
b.indice_frequencia_escolar,
b.populacao,
b.populacao_rural,
b.populacao_urbana,
b.renda_pc,
c.cobertura_fibra,
c.media_cobertura,
c.media_fibra,
c.media_densidade_smp,
c.media_hhi_smp,
c.media_densidade_scm,
c.media_hhi_scm,
d.avg_crescimento,
d.avg_pib,

FROM data_lake_turing_.prata.evolucao_conectividade a
LEFT JOIN data_lake_turing_.prata.idh_uf b ON a.sigla_uf=b.sigla_uf
LEFT JOIN data_lake_turing_.prata.metricas_conectividade_uf c ON a.sigla_uf=c.sigla_uf
LEFT JOIN data_lake_turing_.prata.metricas_pib_uf d ON a.sigla_uf=d.sigla_uf
LIMIT 100
```
